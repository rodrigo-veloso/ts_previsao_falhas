```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('manutencaoexcel.xlsx')
```

```python
df['Data'] = pd.to_datetime(df['Data'], infer_datetime_format=True)
```

```python
df = df[df['Classe']=='CORRETIVA']
```

```python
df['Classe'] = 1
```

```python
df.groupby('Equipamento').count().reset_index().sort_values('Classe',ascending=False).head(10)
```

```python
ts = df[df['Equipamento'] == 'EQP03'].groupby('Data').count().reset_index().drop(["Equipamento"],axis=1)
plt.plot(ts['Data'],ts['Classe'])
```

```python
df['Data'] = df['Data'].apply(lambda x: str(x))
df['Data'] = df['Data'].apply(lambda x: x[:8]+'15'+x[10:])
df['Data'] = df['Data'].astype("datetime64")
```

```python
df['Data'] = df['Data'].apply(lambda x: str(x))
df['Data'] = df['Data'].apply(lambda x: x[:8]+'7'+x[10:] if int(x[8:10])>15 else x[:8]+'22'+x[10:])
df['Data'] = df['Data'].astype("datetime64")
```

```python
ts = df[df['Equipamento'] == 'EQP03'].groupby('Data').count().reset_index().drop(["Equipamento"],axis=1)
plt.plot(ts['Data'],ts['Classe'])
ts.min()
```

```python
rolmean = ts.rolling(window=12).mean()
rolstd = ts.rolling(window=12).std()
plt.plot(rolmean)
plt.plot(rolstd)
plt.plot(ts)
```

```python
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(ts['Classe'],autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-values','Lags Used','Observations'])
for key, value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
dfoutput
```

```python
dif = ts
dif['Classe'] = ts['Classe'] - ts['Classe'].shift()
dif = dif.dropna()
plt.plot(dif['Data'],dif['Classe'])
```

```python
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(dif['Classe'],autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-values','Lags Used','Observations'])
for key, value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
dfoutput
```

```python
ts = ts.set_index(['Data'])
```

```python
ts = ts.dropna()
```

```python
ts.sort_index(inplace=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts,period=10)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
plt.plot(trend)
plt.plot(seasonal)
plt.plot(residual)
```

```python
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(ts, nlags=20)
lag_pacf = pacf(ts, nlags=20, method = 'ols')

```

```python
plt.plot(lag_acf)
plt.plot(np.zeros(20),ls='--')
```

```python
plt.plot(lag_pacf)
plt.plot(np.zeros(20),ls='--')
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(ts,order=(5,1,5))
results_AR = model.fit()
plt.plot(ts)
plt.plot(results_AR.fittedvalues)
print(sum((results_AR.fittedvalues.values-ts['Classe'])**2))
```
